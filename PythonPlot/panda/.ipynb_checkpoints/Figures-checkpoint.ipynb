```python
def timefig(T,a,tit,path,m):
    mpl.rcParams['font.size'] = 20
    plt.figure(figsize=(15, 15)) 
    plt.suptitle(tit,fontsize=20)
    plt.subplot(2,2,1)
    x=T["qf_branch_id"]
    y=T["media_p_total"+m]/60.0
    plt.xlim(np.min(x)-5, np.max(x) + 5)  # Colocamos los límites del eje x
    plt.ylim(np.min(y)-5, np.max(y) + 5)  # Colocamos los límites del eje y
    plt.stem(x[y > y.mean()],y[y > y.mean()], linefmt='k--', markerfmt='go', basefmt='b-',bottom=y.mean(),label="Sobre la media")  # Dibujamos los valores por encima de la media
    plt.stem(x[y < y.mean()],y[y < y.mean()], linefmt='k--', markerfmt='ro', basefmt='b-',bottom=y.mean(),label="Bajo la media")# Dibujamos los valores por debajo de la media
    plt.title(u'Tiempo total de procesos vs Sucursales')  # Colocamos el título del gráfico
    plt.xlabel('qf_branch')  # Colocamos la etiqueta en el eje x
    plt.ylabel('Tiempo [h]')  # Colocamos la etiqueta en el eje y

    plt.subplot(2,2,2)
    x1=T[T["qf_branch_type"]=="Centro de Pagos"]["qf_branch_id"]
    y1=T[T["qf_branch_type"]=="Centro de Pagos"]["media_p_total"+m]/60.0
    plt.xlim(np.min(x)-5, np.max(x) + 5)  # Colocamos los límites del eje x
    plt.ylim(np.min(y)-5, np.max(y) + 5)  # Colocamos los límites del eje y
    if len(y1[y1 < y.mean()])==0:
        plt.stem(x1[y1 > y.mean()],y1[y1 > y.mean()], linefmt='k--', markerfmt='go', basefmt='b--',bottom=y.mean(),label="Sobre la media")  # Dibujamos los valores por encima de la media
    elif len(y1[y1 > y.mean()])==1:
        plt.stem(x1[y1 > y.mean()],y1[y1 > y.mean()], linefmt='k--', markerfmt='go', basefmt='b--',bottom=y.mean(),label="Sobre la media")  # Dibujamos los valores por encima de la media
        plt.stem(x1[y1 < y.mean()],y1[y1 < y.mean()], linefmt='k--', markerfmt='ro', basefmt='b--',bottom=y.mean(),label="Bajo la media")# Dibujamos los valores por debajo de la media
    else:
        plt.stem(x1[y1 < y.mean()],y1[y1 < y.mean()], linefmt='k--', markerfmt='ro', basefmt='b--',bottom=y.mean(),label="Bajo la media")# Dibujamos los valores por debajo de la media
        
    plt.plot(np.linspace(np.min(x)-5,np.max(x) + 5,100),y.mean()*np.ones(100),linewidth=2,color="blue")
    plt.title(u'Tiempo total de procesos vs Sucursales Centros de Pago')  # Colocamos el título del gráfico
    plt.xlabel('qf_branch')  # Colocamos la etiqueta en el eje x
    plt.ylabel('Tiempo [min]')  # Colocamos la etiqueta en el eje y

    plt.subplot(2,2,3)
    x1=T[T["qf_branch_type"]=="Nova"]["qf_branch_id"]
    y1=T[T["qf_branch_type"]=="Nova"]["media_p_total"+m]/60.0
    plt.xlim(np.min(x)-5, np.max(x) + 5)  # Colocamos los límites del eje x
    plt.ylim(np.min(y)-5, np.max(y) + 5)  # Colocamos los límites del eje y
    plt.stem(x1[y1 > y.mean()],y1[y1 > y.mean()], linefmt='k--', markerfmt='go', basefmt='b-',bottom=y.mean(),label="Sobre la media")  # Dibujamos los valores por encima de la media
    plt.stem(x1[y1 < y.mean()],y1[y1 < y.mean()], linefmt='k--', markerfmt='ro', basefmt='b-',bottom=y.mean(),label="Bajo la media")# Dibujamos los valores por debajo de la media
    plt.title(u'Tiempo total de procesos vs Sucursales Nova')  # Colocamos el título del gráfico
    plt.xlabel('qf_branch')  # Colocamos la etiqueta en el eje x
    plt.ylabel('Tiempo [min]')  # Colocamos la etiqueta en el eje y

    plt.subplot(2,2,4)
    x1=T[T["qf_branch_type"]=="Tradicional"]["qf_branch_id"]
    y1=T[T["qf_branch_type"]=="Tradicional"]["media_p_total"+m]/60.0
    plt.xlim(np.min(x)-5, np.max(x) + 5)  # Colocamos los límites del eje x
    plt.ylim(np.min(y)-5, np.max(y) + 5)  # Colocamos los límites del eje y
    plt.stem(x1[y1 > y.mean()],y1[y1 > y.mean()], linefmt='k--', markerfmt='go', basefmt='b-',bottom=y.mean(),label="Sobre la media")  # Dibujamos los valores por encima de la media
    plt.stem(x1[y1 < y.mean()],y1[y1 < y.mean()], linefmt='k--', markerfmt='ro', basefmt='b-',bottom=y.mean(),label="Bajo la media")# Dibujamos los valores por debajo de la media
    plt.title(u'Tiempo total de procesos vs Sucursales Tradicionales')  # Colocamos el título del gráfico
    plt.xlabel('qf_branch')  # Colocamos la etiqueta en el eje x
    plt.ylabel('Tiempo [min]')  # Colocamos la etiqueta en el eje y
    plt.savefig(path+'F'+a+m+'.pdf')
    #plt.show()
```

```python
def acum(T,m):
    n=T.shape[0]
    tot=T["procesos_total"+m].sum()
    p_1=0.0
    p_2=0.0
    p_3=0.0
    p_4=0.0
    for i in range(0,n):
        p_1   +=T["c_internos_s_circular"+m][i]
        p_2   +=T["c_internos_s_lineal"+m][i]
        p_3   +=T["c_externos_s_circular"+m][i]
        p_4   +=T["c_externos_s_lineal"+m][i]
    p_1=p_1*10/tot
    p_2=p_2*10/tot
    p_3=p_3*10/tot
    p_4=p_4*10/tot
    return[p_1,p_2,p_3,p_4]

def piefig(T8p,T8p_0,T8p_3,T8p_4,m,M):    
    plt.figure(figsize=(12, 10))
    #plt.subplot(2,2,1)
    x_list = acum(T8p,m)
    label_list = ["Clientes Internos 1s", "Clientes Internos 2s", 
                  "Clientes Externos 1s","Clientes Externos 2s"]
    plt.axis("equal")
    plt.pie(
            x_list,
            labels=label_list,
            autopct="%.2f%%"
            )
    plt.title("Total de procesos "+M+" realizados por")
    plt.savefig(home+'/DATATON/Image/totclientes1'+m+'.pdf')
    plt.figure(figsize=(12, 10))
    #plt.subplot(2,2,2)
    x_list = acum(T8p_0,m)
    label_list = ["Clientes Internos 1s", "Clientes Internos 2s", 
                  "Clientes Externos 1s","Clientes Externos 2s"]
    plt.axis("equal")
    plt.pie(
            x_list,
            labels=label_list,
            autopct="%.2f%%"
            )
    plt.title("Procesos completos "+M+" realizados por")
    plt.savefig(home+'/DATATON/Image/totclientes2'+m+'.pdf')
    plt.figure(figsize=(12, 10))
    #plt.subplot(2,2,3)
    x_list = acum(T8p_3,m)
    label_list = ["Clientes Internos 1s", "Clientes Internos 2s", 
                  "Clientes Externos 1s","Clientes Externos 2s"]
    plt.axis("equal")
    plt.pie(
            x_list,
            labels=label_list,
            autopct="%.2f%%"
            )
    plt.title("Procesos abandonados "+M+" realizados por")
    plt.savefig(home+'/DATATON/Image/totclientes3'+m+'.pdf')
    #plt.subplot(2,2,4)
    plt.figure(figsize=(12, 10))
    x_list = acum(T8p_4,m)
    label_list = ["Clientes Internos 1s", "Clientes Internos 2s", 
                  "Clientes Externos 1s","Clientes Externos 2s"]
    plt.axis("equal")
    plt.pie(
            x_list,
            labels=label_list,
            autopct="%.2f%%"
            )
    plt.title("Procesos abortados "+M+" realizados por")
    plt.savefig(home+'/DATATON/Image/totclientes4'+m+'.pdf')
    
piefig(T8p,T8p_0,T8p_3,T8p_4,"","") 
#piefig(T8p,T8p_0,T8p_3,T8p_4,"_m8","(Agosto)")
#piefig(T8p,T8p_0,T8p_3,T8p_4,"_m9","(Septiembre)")
#piefig(T8p,T8p_0,T8p_3,T8p_4,"_m10","(Octubre)")
#plt.show()
```

```python
def point(x,y,e,ptext,c):
        plt.annotate("Suc "+str(e), color=c,fontweight='bold',
                     xy = (e, y[x.loc[x==e].index[0]]), 
                     xycoords = 'data', 
                     xytext = (e+ptext[0], 
                     y[x.loc[x==e].index[0]]+ptext[1]), 
                     textcoords = 'data', 
                     bbox=dict(boxstyle="round", fc="none"),
                     arrowprops = dict(arrowstyle = "->",))
def stemplot_up(x,y,lb):
    plt.stem(x[y > y.mean()],y[y > y.mean()], 
             linefmt='k--', markerfmt='go', 
             basefmt='b-',bottom=y.mean(),label=lb)  # Dibujamos los valores por encima de la media
def stemplot_down(x,y,lb):
    plt.stem(x[y < y.mean()],y[y < y.mean()], 
             linefmt='k--', markerfmt='ro', 
             basefmt='b-',bottom=y.mean(),label=lb)# Dibujamos los valores por debajo de la media

cl=("darkgreen","crimson","lightgreen","coral")
mpl.rcParams['font.size'] = 14
plt.figure(figsize=(15, 20)) 
#plt.suptitle("Procesos vs Tiempos medios",fontsize=20)
horast=(31*2+30)*24
x=T8p.sort_values("qf_branch_id")["qf_branch_id"]
e1=x[(T8p.sort_values("qf_branch_id")["procesos_completos"]/(T8t.sort_values("qf_branch_id")["tiempo_p_total"]/3600)).idxmax]
e2=x[(T8p.sort_values("qf_branch_id")["procesos_completos"]/(T8t.sort_values("qf_branch_id")["tiempo_p_total"]/3600)).idxmin]
e3=x[(T8p.sort_values("qf_branch_id")["procesos_completos"]/horast).idxmax]
e4=x[(T8p.sort_values("qf_branch_id")["procesos_completos"]/horast).idxmin]

plt.subplot(3,2,4)
y=T8p.sort_values("qf_branch_id")["procesos_completos"]/(T8t.sort_values("qf_branch_id")["tiempo_p_total"]/3600)
x=T8p.sort_values("qf_branch_id")["qf_branch_id"]
#plt.plot(x,y,linestyle="--",marker="o")
plt.xlim(np.min(x)-5, np.max(x) + 5)  # Colocamos los límites del eje x
plt.ylim(np.min(y)-0.5, np.max(y) + 0.5)  # Colocamos los límites del eje y
stemplot_up(x,y,"")
stemplot_down(x,y,"")
point(x,y,e1,[0,0.3],cl[0])
point(x,y,e2,[100,-0.3],cl[1])
point(x,y,e3,[0,-1],cl[2])
point(x,y,e4,[-100,1.2],cl[3])
plt.title(u'Efectividad por hora invertida vs Sucursal')  # Colocamos el título del gráfico
plt.xlabel('Id Sucursal')  # Colocamos la etiqueta en el eje x
plt.ylabel('Efectividad[#procesos exitosos/horas_invertidas]')  # Colocamos la etiqueta en el eje y

plt.subplot(3,2,6)
y=T8p.sort_values("qf_branch_id")["procesos_completos"]/horast
x=T8p.sort_values("qf_branch_id")["qf_branch_id"]
#plt.plot(x,y,linestyle="--",marker="o")
plt.xlim(np.min(x)-5, np.max(x) + 5)  # Colocamos los límites del eje x
plt.ylim(np.min(y)-0.5, np.max(y) + 0.5)  # Colocamos los límites del eje y
stemplot_up(x,y,"")
stemplot_down(x,y,"")
point(x,y,e1,[100,10],cl[0])
point(x,y,e2,[-100,0],cl[1])
point(x,y,e3,[50,-5],cl[2])
point(x,y,e4,[-100,1],cl[3])
plt.title(u'Efectividad por hora vs Sucursal')  # Colocamos el título del gráfico
plt.xlabel('Id Sucursal')  # Colocamos la etiqueta en el eje x
plt.ylabel('Efectividad[#procesos exitosos/h]')  # Colocamos la etiqueta en el eje y

plt.subplot(3,2,1)
y=T8p.sort_values("qf_branch_id")["procesos_abandonados"]/T8p.sort_values("qf_branch_id")["procesos_completos"]
x=T8p.sort_values("qf_branch_id")["qf_branch_id"]
#plt.plot(x,y,linestyle="--",marker="o")
plt.xlim(np.min(x)-5, np.max(x) + 5)  # Colocamos los límites del eje x
plt.ylim(np.min(y)-0.04, np.max(y) + 0.04)  # Colocamos los límites del eje y
stemplot_up(x,y,"")
stemplot_down(x,y,"")
point(x,y,e1,[-150,-0.03],cl[0])
point(x,y,e2,[-100,0.01],cl[1])
point(x,y,e3,[50,-0.08],cl[2])
point(x,y,e4,[-100,-0.053],cl[3])
plt.title(u'Razón entre procesos abandonados y completados vs Sucursal')  # Colocamos el título del gráfico
plt.ylabel('Razón procesos abandonados/completados')  # Colocamos la etiqueta en el eje x
plt.xlabel('Id Sucursal')  # Colocamos la etiqueta en el eje y

plt.subplot(3,2,2)
y=T8p.sort_values("qf_branch_id")["procesos_completos"]*100/T8p.sort_values("qf_branch_id")["procesos_total"]
x=T8p.sort_values("qf_branch_id")["qf_branch_id"]
#plt.plot(x,y,linestyle="--",marker="o")
plt.xlim(np.min(x)-4, np.max(x) + 4)  # Colocamos los límites del eje x
plt.ylim(np.min(y)-4, np.max(y) + 4)  # Colocamos los límites del eje y
stemplot_up(x,y,"")
stemplot_down(x,y,"")
point(x,y,e1,[100,3],cl[0])
point(x,y,e2,[-100,-3],cl[1])
point(x,y,e3,[-100,8],cl[2])
point(x,y,e4,[-100,5],cl[3])
plt.title(u'Eficacia vs Sucursal')  # Colocamos el título del gráfico
plt.xlabel('Id Sucursal')  # Colocamos la etiqueta en el eje x
plt.ylabel('Eficacia[%]')  # Colocamos la etiqueta en el eje y

plt.subplot(3,2,3)
y=T8p.sort_values("qf_branch_id")["procesos_total"]/(T8t.sort_values("qf_branch_id")["tiempo_p_total"]/3600)
x=T8p.sort_values("qf_branch_id")["qf_branch_id"]
#plt.plot(x,y,linestyle="--",marker="o")
plt.xlim(np.min(x)-5, np.max(x) + 5)  # Colocamos los límites del eje x
plt.ylim(np.min(y)-0.5, np.max(y) + 0.5)  # Colocamos los límites del eje y
stemplot_up(x,y,"")
stemplot_down(x,y,"")
point(x,y,e1,[50,0.4],cl[0])
point(x,y,e2,[-100,-0.4],cl[1])
point(x,y,e3,[-100,1],cl[2])
point(x,y,e4,[-100,1.2],cl[3])
plt.title(u'Eficiencia vs Sucursal')  # Colocamos el título del gráfico
plt.xlabel('Id Sucursal')  # Colocamos la etiqueta en el eje x
plt.ylabel('Eficiencia[#procesos/horas_invertidas]')  # Colocamos la etiqueta en el eje y

plt.subplot(3,2,5)
y=T8p.sort_values("qf_branch_id")["procesos_total"]/horast
x=T8p.sort_values("qf_branch_id")["qf_branch_id"]
#plt.plot(x,y,linestyle="--",marker="o")
plt.xlim(np.min(x)-5, np.max(x) + 5)  # Colocamos los límites del eje x
plt.ylim(np.min(y)-0.5, np.max(y) + 0.5)  # Colocamos los límites del eje y
stemplot_up(x,y,"")
stemplot_down(x,y,"")
point(x,y,e1,[-150,-5],cl[0])
point(x,y,e2,[-100,-5],cl[1])
point(x,y,e3,[70,0],cl[2])
point(x,y,e4,[-100,1],cl[3])
plt.title(u'Eficiencia vs Sucursal')  # Colocamos el título del gráfico
plt.xlabel('Id Sucursal')  # Colocamos la etiqueta en el eje x
plt.ylabel('Eficiencia[#procesos/h]')  # Colocamos la etiqueta en el eje y
plt.savefig(home+'/DATATON/Image/efecty.pdf')
plt.clf()
plt.cla()
plt.close("all") 
#plt.show()
```
